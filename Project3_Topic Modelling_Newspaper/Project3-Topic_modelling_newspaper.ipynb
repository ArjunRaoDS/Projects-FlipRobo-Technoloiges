**Data preprocessing and cleaning**

```python
shear=[i.replace('\xe2\x80\x9c','') for i in documents ]
shear=[i.replace('\xe2\x80\x9d','') for i in shear ]
shear=[i.replace('\xe2\x80\x99s','') for i in shear ]

shears = [x for x in shear if x != ' ']
shearss = [x for x in shears if x != '']
```

```python
dubby=[re.sub("[^a-zA-Z]+", " ", s) for s in shearss]
dubby
```

```python
stemmer = SnowballStemmer("english")
```

```python
# writing function for the entire dataset
def lemmatize_stemming(text):
    return stemmer.stem(WordNetLemmatizer().lemmatize(text,pos='v'))

#Tokenize and Lemmatize
def preprocess(text):
    result=[]
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token)>3:
            result.append(lemmatize_stemming(token))
            
    return result
        
```

```python
processed_docs = []

for doc in dubby:
    processed_docs.append(preprocess(doc))
```

**Creating a dictionary**

```python
#bag of words on the data set
dictionary = gensim.corpora.Dictionary(processed_docs)
```

```python
print (dictionary)
```

```python
#remove rare and repeatative words
dictionary.filter_extremes(no_below=15,no_above=0.1,keep_n=4000)
```

```python
print (dictionary)
```

```python
#Create the Bag-of-words model for each document i.e for each document we create a dictionary reporting how many
#words and how many times those words appear. Save this to 'bow_corpus'
bow_corpus = [dictionary.doc2bow(doc) for doc in processed_docs]
```

```python
#preview 
document_num = 10
bow_doc_x = bow_corpus[document_num]

for i in range(len(bow_doc_x)):
    print("Word {} (\"{}\") appears {} time.".format(bow_doc_x[i][0], 
                                                     dictionary[bow_doc_x[i][0]],
                                                     bow_doc_x[i][1]))
```

**Topic Modelling**

```python
lda_model =  gensim.models.LdaMulticore(bow_corpus, 
                                   num_topics = 25, 
                                   id2word = dictionary,                                    
                                   passes = 10,
                                   workers = 2)
```

```python
for idx, topic in lda_model.print_topics():
    print("Topic: {} \nWords: {}".format(idx, topic ))
    print("\n")
```
